```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y= x**2
```

```python
print(x)
print(x)
```

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("y label")
plt.title("Plot title")
#plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# object oriented method

```python
fig = plt.figure()
axes = fig.add_axes([0.1,.1,.8,.8])
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("y label")
axes.set_title("title")
```

```python
fig = plt.figure()
axis1 = fig.add_axes([.1,.1,.8,.8])
axis1.plot(y,x)
axis1.plot(x,y**x)
axes2 = fig.add_axes([0.3,.3,.3,.3])
axes2.plot(x,y)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([.1,.1,.8,.8])
axis1.plot(y,x)
axis1.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# figure size and DPI

```python
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
